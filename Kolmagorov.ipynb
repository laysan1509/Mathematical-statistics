```python
%matplotlib inline
from scipy import stats
import numpy as np
import pandas as pd
from scipy.stats import kstest
import seaborn
from matplotlib.pyplot import show 
import math
```

```python
data = pd.read_csv("/Users/rakip/Desktop/Course_Work_MS/r2z2.csv")
lambda_ = 1.5
alpha = 0.1
```

```python
from statsmodels.distributions.empirical_distribution import ECDF  #  эмпирическая функция распределения
from scipy.stats import expon
import matplotlib.pyplot as plt

file = ECDF(data['X'])
data['y_values'] = data['X'].apply(file)
#seaborn.lineplot(data['X'],data['y_values'], drawstyle='steps-pre')
seaborn.lineplot(data['X'],data['y_values'], drawstyle='steps-post')

ECDF(data['X'], side='right')
data_exp = np.random.exponential(scale = 1/lambda_, size=7000)
ecdf = ECDF(data_exp)
plt.plot(ecdf.x,ecdf.y)

data_norm = np.random.normal(0,1, size = 3000)
ecdf = ECDF(data_norm)
#plt.plot(ecdf.x,ecdf.y)
```

```python
quant = math.sqrt((-1/2*math.log(alpha/2, math.e)))
print('\nСкрит = Fkolm^(-1)(1 - alpha) = ',quant)
# Критическая область sqrt(n)*D > Cкрит
```

```python
D, p_value = stats.kstest(data['X'], 'expon', args = (0, (1/1.5)))
print('\nKS-statistic D =', D, 'p-value = ', p_value)

if math.sqrt(len(data))*D > quant:
    print("\nT > Cкрит")
    print(math.sqrt(len(data))*D, ' > ', quant)
    print('Н0 отклоняется')
```
